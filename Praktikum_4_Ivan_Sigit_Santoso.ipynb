```python
#Praktikum 1 Perulangan For dengan Range
for i in range(1, 6):
    print(i)
```

```python
#Praktikum 02 - Perulangan while
n = int(input("Masukkan angka: "))
while n > 0:
  print(n)
  n -= 1
```

```python
#Praktikum 03 - Nested Loop (Segitiga Bintang)
for i in range(1, 6):
  for j in range(i):
    print("*", end="")
  print()
```

```python
#Praktikum 04-Break
for i in range(1, 11):
  if i == 5:
   print("Angka 5 ditemukan")
   break
```

```python
#Praktikum 4 - Continue
for i in range (1,6):
  if i % 2 == 1:
      continue
  print(i)
```

```python
#Praktikum 05 - For pada Dictionary
data = {"Nama": "Andi", "Umur": 20, "Prodi": "TRK"}
for k, v in data.items():
    print(k, ":", v)
```

```python
#Praktikum 06 - For denan Enumerate
daftar = ["Andi", "Budi", "Cici"]

for i, nama in enumerate(daftar, start=1):
    print(i, nama)
```

```python
#Praktikum 07 â€“ Studi Kasus: Program Kasir Multi-Item
total = 0

while True:
  barang = input("Masukkan nama barang (ketik 'selesai' untuk  berhenti):")
  if barang.lower() == "selesai":
        break
  harga = float(input(f"Masukkan harga {barang}: "))
  jumlah = int(input(f"Masukkan jumlah {barang}: "))
  subtotal = harga * jumlah
  total += subtotal
print("\n===== Struk Belanja =====")
print(f"Total Belanja : Rp {total}")
```

```python
#Penugasan 01 - Menampilkan Deret Angka
print("Deret kelipatan 3:")
for i in range(3, 31, 3):
    print(i, end=" ")
print()
```

```python
#Penugasan 02 0 Hitung Faktorial
angka = int(input("Masukkan angka untuk faktorial: "))
hasil = 1
counter = angka

while counter > 0:
    hasil *= counter
    counter -= 1

print(f"Faktorial dari {angka} adalah {hasil}")
```

```python
#Penugasan 03 - Pola Segitiga Terbalik (Nedted Loop)
for i in range(6, 0, -1 ):
  for j in range(i):
    print("*", end="")
  print()
```

```python
#Penugasan 04 - Cari Bilangan Prima dengan Break
angka = int(input("Masukkan bilangan untuk cek prima: "))
is_prima = True

if angka < 2:
    is_prima = False
else:
    for i in range(2, angka):
        if angka % i == 0:
            is_prima = False
            break

if is_prima:
    print(f"{angka} adalah Bilangan Prima")
else:
    print(f"{angka} BUKAN Bilangan Prima")
```

```python
#Penugasan 05 - Tampilkan Hanya Bilangan Genap (Continue)
print("Bilangan Genap 1-20:")
for i in range(1, 21):
    if i % 2 != 0:
        continue
    print(i, end=" ")
print()
```

```python
#Penugasan 06 - Literasi Dictionary Mahasiswa
mahasiswa = {"Nama": "Budi", "Umur": 21, "Prodi": "Teknik Komputer"}

print("Data Mahasiswa:")
for k, v in mahasiswa.items():
    print(f"{k} : {v}")
```

```python
#Penugasan 07 - Enumerate Daftar Barang
daftar_barang = ["Buku", "Pulpen", "Tas"]

for i, nama in enumerate(daftar_barang, start=1):
    print(i, nama)
```
