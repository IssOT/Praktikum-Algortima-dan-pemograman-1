```python
#Praktikum 03 - Fungsi dengan Default Argument
def sapa(nama="Mahasiswa"):
    print(f"Halo, {nama}!")

sapa()          # tanpa argumen
sapa("Andi")    # dengan argumen
```

```python
#Praktikum 04 - Fungsi dengn *args dan **kwargs
def jumlahkan(*args):
     return sum(args)
def biodata(**kwargs):
     for key, value in kwargs.items():
       print(f"{key}: {value}")
       print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

```python
#Praktikum 05 - Scope Variabel
x = 10

def ubah():
     x = 5
     print("Nilai lokal:", x)

ubah()
print("Nilai global:", x)
```

```python
#Praktikum 06 - Fungsi Lambda
kali = lambda x, y: x * y
print("Hasil perkalian:", kali(4, 5))
```

```python
#Praktikum 07 - Fungsi Decorator
def decorator_sapa(fungsi):
    def wrapper():
        print("Sebelum fungsi dijalankan...")
        fungsi()
        print("Setelah fungsi dijalankan...")
    return wrapper

@decorator_sapa
def sapa():
    print("Halo, Mahasiswa Polines!")

sapa()
```

```python
#Praktikum 08 - Studi Kasus: Menghitung Rata-rata Nilai Mahasiswa
def hitung_rata(data):
  return sum(data) / len(data)

nilai = [80, 90, 85]
rata = hitung_rata(nilai)
print("Rata-rata nilai:", rata)
```

```python
#Penugasan 02 - Fungsi dengan Parameter dan Return
def luas_persegi(panjang, lebar):
    luas = panjang * lebar
    return luas

print(f"Luas 1: {luas_persegi(10, 5)}")
print(f"Luas 2: {luas_persegi(4, 4)}")
print(f"Luas 3: {luas_persegi(7, 3)}")
```

```python
#Penugasan 03 - Fungsi dengan Default Argument
def salam(waktu="pagi", nama="Mahasiswa"):
    print(f"Selamat {waktu}, {nama}!")
salam()
salam("siang")
salam("malam", "Budi")
```

```python
#Penugasan 04 - Fungsi dengan *args
def jumlahkan_semua(*angka):
    total = sum(angka)
    return total

print(f"Total (3 angka): {jumlahkan_semua(10, 20, 30)}")
print(f"Total (5 angka): {jumlahkan_semua(1, 2, 3, 4, 5)}")
print(f"Total (7 angka): {jumlahkan_semua(100, 200, 300, 400, 500, 600, 700)}")
```

```python
#Penugasan 05 - Fungsi dengan **kwargs
def info_barang(**data):
    print("--- Info Barang ---")
    for kunci, nilai in data.items():
        print(f"{kunci}: {nilai}")

info_barang(Nama="Buku", Harga=15000, Stok=30)
info_barang(Nama="Laptop", Merk="Asus", Processor="i7", RAM="16GB")
```

```python
#Penugasan 06 - Scope Variabel
x = 100  # Variabel Global

def ubah():
    x = 50  # Variabel Lokal (hanya hidup di dalam fungsi ini)
    print(f"Di dalam fungsi: {x}")

ubah()
print(f"Di luar fungsi: {x}")

# Penjelasan:
# Output di luar tetap 100 karena x di dalam fungsi 'ubah'
# dianggap sebagai variabel baru yang berbeda alamat mematinya.
```

```python
#Penugasan 07 - Fungsi Lambda
import math
kuadrat = lambda x: x ** 2
keliling = lambda r: 2 * math.pi * r

print(f"Kuadrat dari 5 : {kuadrat(5)}")
print(f"Keliling r=7   : {keliling(7):.2f}")
```

```python
#Penugasan 08 - Fungsi Decorator
def logger(func):
    def wrapper():
        print("Memulai fungsi...")
        func()
        print("Fungsi selesai dijalankan.")
    return wrapper

@logger
def halo():
    print("Halo semuanya!")
halo()
```

```python
#Penugasan 09 - Studi Kasus Nyata - Program Nilai
def input_nilai():
    data = input("Masukkan nilai mahasiswa (pisahkan dengan koma, cth: 80,90,75): ")
    return [int(x) for x in data.split(',')]

def hitung_rata(daftar_nilai):
    return sum(daftar_nilai) / len(daftar_nilai)

def status_kelulusan(rata):
    if rata >= 60:
        return "Lulus"
    else:
        return "Tidak Lulus"

hitung_akhir = lambda tugas, ujian: (tugas * 0.40) + (ujian * 0.60)

print("=== PROGRAM NILAI MAHASISWA ===")
nilai_list = input_nilai()
rata_rata = hitung_rata(nilai_list)
status = status_kelulusan(rata_rata)
print(f"Rata-rata nilai : {rata_rata:.2f}")
print(f"Status          : {status}")
print("\n--- Hitung Nilai Bobot ---")
t = int(input("Nilai Tugas : "))
u = int(input("Nilai Ujian : "))
akhir = hitung_akhir(t, u)
print(f"Nilai Akhir (Bobot) : {akhir:.2f}")
```
