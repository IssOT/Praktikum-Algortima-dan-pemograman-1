```python
#Praktikum 01 IF Sederhana
nilai = int(input("Masukkan nilai ujian: "))
if nilai >= 60:
  print("Anda LULUS")
```

```python
#Praktikum 02 IF ELSE
nilai = int(input("Masukkan nilai ujian: "))
if nilai >= 60:
    print("Anda LULUS")
else:
     print("Anda TIDAK LULUS")
```

```python
#Praktikum 03 IF ELIF ELSE (Grade)
nilai = int(input("Masukkan nilai akhir: "))

if nilai >= 85:
    grade = "A"
elif nilai >= 70:
    grade = "B"
elif nilai >= 55:
     grade = "C"
elif nilai >= 40:
     grade = "D"
else:
     grade = "E"

print(f"Nilai Anda {nilai}, maka grade Anda = {grade}")
```

```python
#Praktikum 04 Nested If
ipk = float(input("Masukkan IPK: "))
sks = int(input("Masukkan jumlah SKS: "))

if ipk >= 3.0:
  if sks >= 18:
   print("Anda LAYAK mendapat beasiswa")
  else:
   print("IPK memenuhi, tapi SKS kurang")
else:
  print("Anda TIDAK layak mendapat beasiswa")
```

```python
#Praktikum 05 Match Case
menu = int(input("Pilih menu (1-3): "))

match menu:
     case 1:
        print("Anda memilih Nasi Goreng")
     case 2:
        print("Anda memilih Mie Ayam")
     case 3:
         print("Anda memilih Soto Ayam")
     case _:
         print("Menu tidak tersedia")
```

```python
#Praktikum 06 Studi Kasus: Program Kasir
barang = input("Masukkan nama barang: ")
harga = float(input("Masukkan harga barang: "))
jumlah = int(input("Masukkan jumlah barang: "))

total = harga * jumlah
if total >= 500000:
     diskon = 0.2
elif total >= 250000:
     diskon = 0.1
else:
     diskon = 0
bayar = total - (total * diskon)
print("\n===== Struk Belanja =====")
print(f"Barang : {barang}")
print(f"Jumlah : {jumlah}")
print(f"Total  : Rp {total}")
print(f"Diskon : {diskon*100}%")
print(f"Bayar  : Rp {bayar}")
```

```python
#Penugasan Bilangan Positif/Negatif/Nol
bilangan = int(input("Masukkan sebuah bilangan bulat: "))

if bilangan > 0:
    print("Bilangan tersebut adalah Positif.")
elif bilangan < 0:
    print("Bilangan tersebut adalah Negatif.")
else:
    print("Bilangan tersebut adalah Nol.")
```

```python
#Menentukan Kategori Usia
umur = int(input("Masukkan umur Anda: "))

if umur < 5:
    kategori = "Balita"
elif 5 <= umur <= 11:
    kategori = "Anak-anak"
elif 12 <= umur <= 17:
    kategori = "Remaja"
elif 18 <= umur <= 59:
    kategori = "Dewasa"
else:
    kategori = "Lansia"

print(f"Kategori usia: {kategori}")
```

```python
#Hitung Tarif Parkir
durasi = int(input("Lama parkir (jam): "))

if durasi <= 2:
    tarif = 5000
else:
    # 5000 untuk 2 jam pertama + 2000 untuk sisa jam
    tarif = 5000 + ((durasi - 2) * 2000)

# Cek tarif maksimal
if tarif > 20000:
    tarif = 20000

print(f"Total tarif parkir: Rp {tarif:,}")
```

```python
#One-Liner If (Cek Ganjil/Genap)
angka = int(input("Masukkan angka: "))
status = "Genap" if angka % 2 == 0 else "Ganjil"

print(f"Angka {angka} adalah bilangan {status}.")
```

```python
#Menu Restoran dengan Match-Case
print("=== MENU MAKANAN ===")
print("1. Nasi Goreng")
print("2. Mie Ayam")
print("3. Soto Ayam")

pilihan = int(input("Masukkan pilihan menu (1-3): "))

match pilihan:
    case 1:
        print("Anda memesan: Nasi Goreng")
    case 2:
        print("Anda memesan: Mie Ayam")
    case 3:
        print("Anda memesan: Soto Ayam")
    case _:
        print("Menu tidak tersedia")
```

```python
#Studi Kasus Nyata - Program Penentu Diskon
print("=== KASIR KOPMA ELEKTRO ===")
nama_barang = input("Nama Barang  : ")
jumlah = int(input("Jumlah Beli  : "))
harga = int(input("Harga Satuan : Rp "))
total_harga = jumlah * harga
diskon = 0

if total_harga >= 500000:
    diskon = total_harga * 0.20  # 20%
elif total_harga >= 250000:
    diskon = total_harga * 0.10  # 10%

total_bayar = total_harga - diskon

print("\n" + "="*30)
print("      STRUK BELANJA      ")
print("="*30)
print(f"Barang      : {nama_barang}")
print(f"Jumlah      : {jumlah}")
print(f"Harga/pcs   : Rp {harga:,}")
print("-" * 30)
print(f"Total Harga : Rp {total_harga:,.0f}")
print(f"Diskon      : Rp {diskon:,.0f}")
print(f"Total Bayar : Rp {total_bayar:,.0f}")
print("="*30)
```
