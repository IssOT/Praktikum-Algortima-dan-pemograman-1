```python
#Praktikum 1 Bubble Sort
def bubble_sort(arr):
   n = len(arr)
   for i in range(n):
      for j in range(0, n-i-1):
          if arr[j] > arr[j+1]:
              arr[j], arr[j+1] = arr[j+1], arr[j]
   return arr
data = [64, 34, 25, 12, 22, 11, 90]
print(bubble_sort(data))
```

```python
#Praktikum 2 - Selection sort
def selection_sort(arr):
 n = len(arr)
 for i in range(n):
      min_index = i
      for j in range(i+1, n):
          if arr[j] < arr[min_index]:
            min_index = j
      arr[i], arr[min_index] = arr[min_index], arr[i]
      return arr

data = [64, 25, 12, 22, 11]
print(selection_sort(data))
```

```python
#Praktikum 3 - Insertion Sort
def insertion_sort(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key  # Corrected: This line should be outside the while loop
    return data  # Corrected: This line should be outside the for loop

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

```python
#Praktikum 04 – Studi Kasus: Mengurutkan Nilai Mahasiswa
def bubble_sort_nilai(data):
     n = len(data)
     for i in range(n):
         for j in range(0, n - i - 1):
             if data[j][1] < data[j + 1][1]:
                 data[j], data[j + 1] = data[j + 1], data[j]
     return data

mahasiswa = [("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

```python
#Praktikum 06 – Sorting Data Barang (Nama & Harga)
def bubble_sort_key(data, key_index=0, reverse=False):
     n = len(data)
     for i in range(n):
         for j in range(0, n - i - 1):
             if reverse:
                 if data[j][key_index] < data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
             else:
                 if data[j][key_index] > data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
     return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

```python
#Penugasan 1 Eksperimen Bubble Sort Descending
def bubble_sort_descending(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            # Ubah '>' menjadi '<' untuk descending
            if arr[j] < arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

data = [64, 34, 25, 12, 22, 11, 90]
print("Bubble Sort Descending:", bubble_sort_descending(data))
```

```python
#Penugasan 2 Eksperimen Selection Sort Max to End
def selection_sort_max_to_end(arr):
    n = len(arr)
    for i in range(n):
        # Asumsikan elemen pertama adalah max
        max_idx = 0
        # Loop hanya sampai n - i (area yang belum terurut)
        for j in range(1, n - i):
            if arr[j] > arr[max_idx]:
                max_idx = j

        # Tukar elemen terbesar dengan elemen terakhir di area belum terurut
        # Posisi tujuan adalah (n - 1 - i)
        target_idx = n - 1 - i
        arr[target_idx], arr[max_idx] = arr[max_idx], arr[target_idx]
    return arr

data = [29, 10, 14, 37, 13]
print("Selection Sort (Max ke Akhir):", selection_sort_max_to_end(data))
```

```python
#Penugasan 3 - Eksperimen Insertion Sort
def insertion_sort_strings(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

mahasiswa = ["Budi", "Andi", "Citra", "Eka", "Dian"]
print("Insertion Sort (String):", insertion_sort_strings(mahasiswa))
```

```python
#Penugasan 5 - Analisis Kompleksitas
import random

# Versi modifikasi untuk menghitung perbandingan
def count_bubble(arr):
    comparisons = 0
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            comparisons += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return comparisons

def count_selection(arr):
    comparisons = 0
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            comparisons += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return comparisons

def count_insertion(arr):
    comparisons = 0
    arr = arr.copy()
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0:
            comparisons += 1 # Menghitung cek kondisi loop
            if key < arr[j]:
                arr[j + 1] = arr[j]
                j -= 1
            else:
                break
        arr[j + 1] = key
    return comparisons

# Pengujian
sizes = [10, 50, 100]
print(f"{'Ukuran':<10} | {'Bubble':<10} | {'Selection':<10} | {'Insertion':<10}")
print("-" * 46)

for size in sizes:
    data = [random.randint(1, 1000) for _ in range(size)]

    c_bubble = count_bubble(data)
    c_select = count_selection(data)
    c_insert = count_insertion(data)

    print(f"{size:<10} | {c_bubble:<10} | {c_select:<10} | {c_insert:<10}")
```

```python
#penugasn 6 - Proyek Mini Mnnjemen barang
def bubble_sort_barang(daftar, key, descending=False):
    n = len(daftar)
    data = daftar.copy()
    for i in range(n):
        for j in range(0, n - i - 1):
            # Kondisi dinamis berdasarkan key ('nama' atau 'harga')
            kondisi = data[j][key] > data[j + 1][key]

            if descending:
                kondisi = data[j][key] < data[j + 1][key]

            if kondisi:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

# Data Awal
barang = [
    {"nama": "Pensil", "harga": 2000},
    {"nama": "Buku", "harga": 5000},
    {"nama": "Penghapus", "harga": 1500},
    {"nama": "Penggaris", "harga": 3000}
]

def main():
    while True:
        print("\n--- Manajemen Barang ---")
        print("1. Tampil Barang")
        print("2. Urutkan berdasarkan Nama (A-Z)")
        print("3. Urutkan berdasarkan Harga (Termurah-Termahal)")
        print("4. Urutkan berdasarkan Harga (Termahal-Termurah)")
        print("5. Keluar")

        pilihan = input("Pilih menu: ")

        if pilihan == '1':
            print(barang)
        elif pilihan == '2':
            hasil = bubble_sort_barang(barang, 'nama', descending=False)
            print("Hasil:", hasil)
        elif pilihan == '3':
            hasil = bubble_sort_barang(barang, 'harga', descending=False)
            print("Hasil:", hasil)
        elif pilihan == '4':
            hasil = bubble_sort_barang(barang, 'harga', descending=True)
            print("Hasil:", hasil)
        elif pilihan == '5':
            break

if __name__ == "__main__":
    main()
```
