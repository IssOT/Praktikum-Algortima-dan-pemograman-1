```python
#Penugasan 01 - Eksperimen dengan Modul math
import math

def hitung_luas_lingkaran(jari_jari):
    """Menghitung luas lingkaran menggunakan math.pi."""
    luas = math.pi * (jari_jari ** 2)
    return luas

# Uji dengan jari-jari 7
luas_7 = hitung_luas_lingkaran(7)
print(f"Luas lingkaran dengan r=7 adalah {luas_7}")

# Uji dengan jari-jari 10
luas_10 = hitung_luas_lingkaran(10)
print(f"Luas lingkaran dengan r=10 adalah {luas_10}")
```

```python
#Penugasan 04 - Modul Buatan Sendiri
%%writefile konversi.py

def celcius_ke_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_ke_celcius(f):

    return (f - 32) * 5/9
```

```python
#Penugasan 04 - Eksekusi
# Mengimpor modul konversi.py yang tadi kita buat
import konversi

print("--- Menggunakan Modul Konversi ---")

suhu_c = 100
suhu_f = 212

# Menggunakan fungsi dari modul
hasil_f = konversi.celcius_ke_fahrenheit(suhu_c)
hasil_c = konversi.fahrenheit_ke_celcius(suhu_f)

print(f"{suhu_c} Celcius adalah {hasil_f} Fahrenheit")
print(f"{suhu_f} Fahrenheit adalah {hasil_c} Celcius")
```

```python
#Penugasan 07 - Modul Studi Kasus Mini - Kuis Random
%%writefile laporan.py

def cetak_laporan(skor, total_soal):
    print("\n--- KUIS SELESAI ---")
    print(f"Skor Anda: {skor} dari {total_soal} soal.")

    # Hitung persentase
    persentase = (skor / total_soal) * 100
    print(f"Nilai Akhir: {persentase:.1f}")

    if persentase == 100:
        print("Sempurna! Kerja bagus!")
    elif persentase >= 70:
        print("Bagus, Anda berhasil!")
    else:
        print("Teruslah berlatih!")
```

```python
#Penugasan 07 - Kuis Random
import random
import laporan

print("=== KUIS PENJUMLAHAN SEDERHANA ===")
print("Jawablah 3 soal berikut:")

skor = 0
total_soal = 3

for i in range(total_soal):
    a = random.randint(1, 20)
    b = random.randint(1, 20)
    jawaban_benar = a + b

    try:
        jawaban_user = int(input(f"Soal {i+1}: {a} + {b} = "))

        if jawaban_user == jawaban_benar:
            print("Benar!")
            skor += 1
        else:
            print(f"Salah. Jawaban yang benar adalah {jawaban_benar}")

    except ValueError:
        print("Input tidak valid. Soal ini dianggap salah.")

laporan.cetak_laporan(skor, total_soal)
```
