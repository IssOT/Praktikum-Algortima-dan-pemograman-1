```python
#Penugasan 05 - Operasi dalam F-String
a = int(input("Masukkan a: "))
b = int(input("Masukkan : "))

print(f"Hasil Penjumlahan : {a} + {b} = {a + b}")
print(f"Hasil Perkalian   : {a} x {b} = {a * b}")
print(f"Hasil Pembagian   : {a} / {b} = {a / b:.3f}")
```

```python
#Penugasan 06 - Casting Data
data_input = input("Masukkan sebuah angka: ")
angka_int = int(data_input)
kuadrat = angka_int ** 2
print(f"Hasil kuadrat (int): {kuadrat}")

angka_float = float(data_input)
print(f"Format desimal (float): {angka_float:.2f}")
```

```python
#Penugasan 07 - Studi Kasus - Analisis Teks
kalimat = input("Tuliskan sebuah kalimat: ")
daftar_kata = kalimat.split()
jumlah_kata = len(daftar_kata)
kata_terpanjang = max(daftar_kata, key=len)
daftar_kata.sort(key=str.lower)

print("\n=== HASIL ANALISIS ===")
print(f"Jumlah kata     : {jumlah_kata}")
print(f"Kata terpanjang : {kata_terpanjang}")
print(f"Urutan abjad    : {daftar_kata}")
```
